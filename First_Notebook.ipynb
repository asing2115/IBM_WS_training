# pseudo code to print the word in triangle pattern

```python
#Below method is define to identify the space in word and throw error
def check_space(s):
    for i in range(len(s)):
        if s[i] == ' ':
            return """Error! Space between words not allowed
            Replace ' ' with '_'"""
```

```python
word = input("Enter a word \n")
s = ''
#check for spaces
found = check_space(word)

if(not(isinstance(found, str))):
    for x in range(len(word)):
        s = s + word[x]
        print(s)
else:
    print(found)

        
```
